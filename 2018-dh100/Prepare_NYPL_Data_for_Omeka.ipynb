```python
import pandas as pd
```

```python
collections = pd.read_csv('data-and-utilities/collections/pd_collections.csv')
```

```python
collections.head()
```

```python
files = ['data-and-utilities/items/pd_items_1.csv', 'data-and-utilities/items/pd_items_2.csv']
frames = [pd.read_csv(f) for f in files]
items = pd.concat(frames, ignore_index=True)
```

```python
items.head()
```

```python
collection_name = 'Maps of North America'
collections[collections['Title'].str.contains(collection_name)]
```

```python
uuid = collections[collections['Title'].str.contains(collection_name)]['UUID'].values
```

```python
collection_items = items[items['Collection UUID'] == uuid[0]]
```
